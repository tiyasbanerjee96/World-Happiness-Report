# Data Visualization

### Plot a graph Positive affect Vs quantity

```python
plt.bar(data['year'],data['Life Ladder'])
plt.title('year Vs Life Ladder',fontsize=15,color='red')
plt.xlabel('year',fontsize=15)
plt.ylabel('Life Ladder',fontsize=15)
plt.grid(True)
plt.show()
```

```python
data.sort_values(by=['Generosity'])
```

### Maximum of Life Ladder

```python
data['Life Ladder'].max()
```

### Minimum of Life Ladder

```python
data['Life Ladder'].min()
```

```python
data1=data.copy()
data1=data1.pivot_table('year',columns='Healthy life expectancy at birth',aggfunc='sum')
data1
```

```python
data2=data.copy()
data2=data2.pivot_table('Country name',columns='Log GDP per capita',aggfunc='sum')
data2
```

```python
data3=data.copy()
data3=data3.pivot_table('Country name',columns='Freedom to make life choices',aggfunc='sum')
data3
```

```python
plt.bar(data['year'],data['Healthy life expectancy at birth'])
plt.title('year Vs Healthy life expectancy at birth',fontsize=15,color='blue')
plt.xlabel('year',fontsize=15)
plt.ylabel('Healthy life expectancy at birth',fontsize=15)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(16,8))
sns.relplot(x='year',y='Generosity',color='green',data=data)
plt.show()
```

```python
sns.countplot(x='year',data=data,palette='coolwarm')
plt.show()
```

```python
plt.title('year',fontsize=20)
data['year'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True)
plt.show()
```

```python
sns.countplot(x='year',data=data,palette='tab10')
plt.show()
```

```python
data.groupby('year')[['Life Ladder','Social support']].sum().plot.bar(color=['red','green'],figsize=(8,5))
plt.ylabel('Life Ladder and Social support')
plt.show()
```

```python
data.hist(figsize=(13,13),layout=(3,3),sharex=False)
plt.show()
```

```python
data.plot(kind='box',figsize=(13,13),layout=(4,4),subplots=True,sharex=False)
plt.show()
```

```python
sns.pairplot(data)
plt.show()
```
